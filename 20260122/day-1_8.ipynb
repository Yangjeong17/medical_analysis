### 1.1. 프로젝트 목표

- 시나리오: 당신은 세계보건기구(WHO)의 데이터 분석가이다.
  개발도상국의 기대 수명(Life Expectancy)을 높이기 위해, 어떤
  사회적/경제적 요인을 우선적으로 개선해야 하는지 정책 제언을 해야 한다.
- 학습 목표:
  1.  데이터의 결측치(Missing Value)를 탐지하고 적절한 값(평균
      등)으로 채운다.
  2.  GDP(수만 달러)와 BMI(수십)처럼 스케일 차이가 큰 데이터를
      표준화한다.
  3.  상관관계 히트맵을 그려 변수 간의 관계를 사전에 파악한다.
  4.  회귀 모델을 통해 기대 수명에 가장 긍정적/부정적 영향을 주는 요인을
      도출한다.

## 2. 프로젝트 개발 가이드 (Step-by-Step)

### Step 1. 데이터 탐색 및 상관관계 분석

- `.isnull().sum()`을 사용하여 결측치가 얼마나 있는지 확인한다.
- `sns.heatmap(df.corr())`을 그려서 기대 수명(Target)과 색깔이
  진하게(상관관계 높게) 표시되는 변수가 무엇인지 눈으로 확인한다.

### Step 2. 결측치 처리 (Imputation)

- 현실 데이터는 항상 비어있는 곳이 있다. GDP나 알코올 섭취량 등이
  비어있는 경우, 해당 컬럼의 평균값(Mean)으로 빈칸을 채운다.
  (`fillna` 함수 활용)

### Step 3. 데이터 분할 및 스케일링

- Target(`Life_Expectancy`)을 분리하고 Train/Test 셋으로 나눈다.
- GDP는 숫자가 매우 크고, Alcohol은 숫자가 작다. `StandardScaler`를
  이용해 단위를 통일한다.

### Step 4. 모델링 및 정책 도출

- `LinearRegression` 모델을 학습시킨다.
- 모델의 계수(Coefficient)를 확인하여 리포트를 작성한다.
  - *질문:* 교육(Schooling) 기간이 1년 늘어나면 기대 수명은 몇 년
    늘어나는가?
  - *질문:* 알코올 섭취는 기대 수명에 긍정적인가 부정적인가?
