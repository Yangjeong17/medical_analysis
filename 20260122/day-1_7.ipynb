## 1. 프로젝트 개요 및 데이터 정의

### 1.1. 프로젝트 목표

- 시나리오: 당신은 보험사의 AI 연구원이다. 신규 가입자의 기본 건강
  정보를 바탕으로, 이 사람이 연간 얼마의 의료비(Charges)를 청구할지
  예측하는 모델을 개발해야 한다.
- 학습 목표:
  1.  문자열 데이터(성별, 흡연 등)를 숫자로 변환한다 (전처리).
  2.  나이와 BMI 등 단위가 다른 데이터를 스케일링한다 (전처리).
  3.  여러 변수를 종합하여 예측 모델을 만든다 (다중 선형 회귀).
  4.  예측 결과의 오차를 계산하고, 의료비에 가장 큰 영향을 미치는 요인을
      찾아낸다 (평가 및 해석).

## 2. 프로젝트 개발 가이드 (Step-by-Step)

학생들이 스스로 고민하며 해결할 수 있도록 단계별 미션을 제시한다.

### Step 1. 데이터 탐색 (EDA)

- 데이터를 생성하고 `.info()`와 `.describe()`로 구조를 파악한다.
- 질문: 흡연자(smoker=‘yes’)와 비흡연자(smoker=‘no’)의 의료비 평균이
  얼마나 차이 나는지 Boxplot으로 시각화하여 확인한다.

### Step 2. 데이터 전처리 (Preprocessing) - 핵심

- 인코딩: `sex`, `smoker`, `region` 컬럼은 문자열이다. 이를 컴퓨터가
  이해할 수 있는 숫자로 바꾼다. (Pandas `get_dummies` 또는 Scikit-learn
  `OneHotEncoder` 활용)
- 데이터 분할: 전체 데이터를 학습용(Train)과 테스트용(Test)으로 8:2
  비율로 나눈다.
- 스케일링: `age`, `bmi` 등의 수치형 변수에 `StandardScaler`를
  적용한다. (주의: Train 데이터 기준으로 Test 데이터를 변환할 것)

### Step 3. 모델링 및 학습 (Modeling)

- `LinearRegression` 모델을 생성하고 학습 데이터로 학습시킨다.
- 추가 미션: 과적합이 의심된다면 `Ridge` 모델도 시도해본다.

### Step 4. 평가 및 해석 (Evaluation)

- 테스트 데이터에 대한 예측을 수행한다.
- RMSE (평균 오차 금액)와 R2 Score (설명력)를 출력한다.
- 회귀 계수 확인: 어떤 변수가 의료비 상승에 가장 큰 영향을 주었는가?
  (흡연 여부가 가장 높게 나와야 정상)
