*   Ïã¨Ìïú Í≤ΩÏö∞ $w_1 = 100, w_2 = -90$ Ï≤òÎüº ÎßêÎèÑ Ïïà ÎêòÎäî Ïà´ÏûêÎ•º Î∂ÄÏó¨Ìï¥ÏÑú Ìï©Í≥ÑÎßå ÎßûÏ∂îÎ†§Í≥† ÌïúÎã§.

# üìò \[3ÍµêÏãú\] Îã§Ï§ë ÏÑ†Ìòï ÌöåÍ∑Ä (Multiple Linear Regression) Ï†ïÎ¶¨

## 3.1. \[Í∞úÎÖê Ïã¨Ìôî\] Î≥µÏû°Ìïú ÌòÑÏã§ÏùÑ ÏÑ§Î™ÖÌïòÎäî Ï∞®ÏõêÏùò ÌôïÏû•

### 1) Ïôú Î≥ÄÏàò ÌïòÎÇòÎ°úÎäî Î∂ÄÏ°±ÌïúÍ∞Ä?

2ÍµêÏãúÏóêÏÑú Ïö∞Î¶¨Îäî `BMI` ÌïòÎÇòÎßåÏúºÎ°ú ÎãπÎá®Î≥ëÏùÑ ÏòàÏ∏°ÌïòÎ†§ ÌñàÎã§. ÌïòÏßÄÎßå ÌòÑÏã§ÏùÄ Í∑∏Î†áÍ≤å Îã®ÏàúÌïòÏßÄ ÏïäÎã§. \* ÏöîÎ¶¨ ÎπÑÏú†: ÍπÄÏπòÏ∞åÍ∞ú ÎßõÏùÑ ÏòàÏ∏°ÌïúÎã§Í≥† Ìï¥Î≥¥Ïûê.
\* *Îã®Ïàú ÌöåÍ∑Ä:* ‚ÄúÏÜåÍ∏à Ïñë($x$)‚ÄùÎßå ÏïåÎ©¥ Îßõ($y$)ÏùÑ Ïïå Ïàò ÏûàÎã§?
$\rightarrow$ Î∂àÍ∞ÄÎä•ÌïòÎã§. ÏÑ§ÌÉï, Í≥†Ï∂ßÍ∞ÄÎ£®, ÍπÄÏπòÏùò Ïã†Îßõ, ÎÅìÏù∏ ÏãúÍ∞Ñ Îì±Ïù¥ Î™®Îëê Ìï©Ï≥êÏ†∏Ïïº ÎßõÏù¥ Í≤∞Ï†ïÎêúÎã§. \* *Îã§Ï§ë ÌöåÍ∑Ä:* Ïó¨Îü¨
Ïû¨Î£å($x_1, x_2, \dots$)Îì§Ïù¥ Í∞ÅÍ∞ÅÏùò ÎπÑÏú®($w_1, w_2, \dots$)Î°ú ÏÑûÏó¨ÏÑú ÏµúÏ¢Ö Í≤∞Í≥º($y$)Î•º ÎßåÎì†Îã§.

### 2) ÏàòÏãùÏùò ÌôïÏû•: ÏÑ†(Line)ÏóêÏÑú Î©¥(Plane)ÏúºÎ°ú

- Îã®Ïàú ÌöåÍ∑Ä ÏàòÏãù: $y = w \cdot x + b$
- Îã§Ï§ë ÌöåÍ∑Ä ÏàòÏãù:
  $y = w_1x_1 + w_2x_2 + w_3x_3 + \dots + w_nx_n + b$
  - ÏûÖÎ†• Î≥ÄÏàò($x$)Í∞Ä ÎäòÏñ¥ÎÇú ÎßåÌÅº, Í∞Å Î≥ÄÏàòÏùò Ï§ëÏöîÎèÑÎ•º ÎÇòÌÉÄÎÇ¥Îäî Í∞ÄÏ§ëÏπò($w$)ÎèÑ ÎäòÏñ¥ÎÇúÎã§.
  - Í∏∞ÌïòÌïôÏ†Å ÏùòÎØ∏: Î≥ÄÏàòÍ∞Ä 2Í∞ú($x_1, x_2$)Í∞Ä ÎêòÎ©¥, Ïö∞Î¶¨Îäî 3Ï∞®Ïõê Í≥µÍ∞ÑÏóêÏÑú Îç∞Ïù¥ÌÑ∞Î•º Í∞ÄÏû• Ïûò ÏÑ§Î™ÖÌïòÎäî ‚ÄòÌèâÎ©¥(Plane)‚ÄôÏùÑ Ìé¥Îäî Í≤ÉÏù¥ÎêúÎã§. Î≥ÄÏàòÍ∞Ä Í∑∏ Ïù¥ÏÉÅÏù¥Î©¥ 4Ï∞®Ïõê Ïù¥ÏÉÅÏùò ‚ÄòÏ¥àÌèâÎ©¥(Hyperplane)‚ÄôÏù¥ÎêúÎã§.

### 3) Ìó¨Ïä§ÏºÄÏñ¥ÏóêÏÑúÏùò ‚ÄòÍ∞ÄÏ§ëÏπò($w$)‚Äô Ìï¥ÏÑù (Ï§ëÏöî)

ÏùòÏÇ¨Îì§ÏùÄ ÏòàÏ∏°Í∞í ÏûêÏ≤¥Î≥¥Îã§ $w$Í∞íÏùÑ Îçî Ï§ëÏöîÌïòÍ≤å Î≥º ÎïåÍ∞Ä ÎßéÎã§. Ïù¥Î•º Î∂ÄÎ∂Ñ ÌöåÍ∑Ä Í≥ÑÏàò(Partial Regression Coefficient)ÎùºÍ≥† ÌïúÎã§. \* ÏùòÎØ∏: ‚ÄúÎã§Î•∏Î™®Îì† Î≥ÄÏàòÍ∞Ä Í≥†Ï†ïÎêòÏñ¥ ÏûàÎã§Í≥† Í∞ÄÏ†ïÌï† Îïå($ceteris paribus$), $x_1$Ïù¥ 1Îã®ÏúÑ Ï¶ùÍ∞ÄÌïòÎ©¥ $y$Îäî $w_1$ÎßåÌÅº Î≥ÄÌïúÎã§.‚Äù \* *ÏòàÏãú:* Ïã¨Ïû•Î≥ë ÏúÑÌóòÎèÑ = $0.5 \times \text{ÎÇòÏù¥} + 2.0 \times \text{Ìù°Ïó∞Ïó¨Î∂Ä} + \dots$ \* Ïó¨Í∏∞ÏÑú $w_{Ìù°Ïó∞}=2.0$Ïùò ÏùòÎØ∏Îäî, Í∞ôÏùÄ ÎÇòÏù¥, Í∞ôÏùÄ ÏÑ±Î≥Ñ, Í∞ôÏùÄ ÌòàÏïïÏùÑ Í∞ÄÏßÑ ÏÇ¨ÎûåÏù¥ÎùºÎèÑ Ìù°Ïó∞ÏùÑ ÌïòÎ©¥ Ïã¨Ïû•Î≥ë ÏúÑÌóòÎèÑÍ∞Ä 2.0ÎßåÌÅº Îõ¥Îã§Îäî ÎúªÏù¥Îã§. Ï¶â, ÏßàÎ≥ëÏùò ‚ÄòÏúÑÌóò Ïù∏Ïûê(Risk Factor)‚ÄôÎ•º Ï∞æÏïÑÎÇ¥Îäî Í≥ºÏ†ïÏù¥Îã§.

### 4) ÏöîÏïΩ: ÌïµÏã¨ Î©îÏãúÏßÄ

1.  Îã§Ï§ë ÏÑ†Ìòï ÌöåÍ∑Ä: ÌòÑÏã§ ÏÑ∏Í≥ÑÏùò Î≥µÏû°Ìïú Ïù∏Í≥ºÍ¥ÄÍ≥ÑÎ•º ÏÑ§Î™ÖÌïòÍ∏∞ ÏúÑÌï¥ ÏûÖÎ†• Î≥ÄÏàò($x$)Î•º Ïó¨Îü¨ Í∞úÎ°ú ÌôïÏû•Ìïú Î™®Îç∏Ïù¥Îã§. ($y = w_1x_1 + \dots + w_nx_n + b$)
2.  Í∏∞ÌïòÌïôÏ†Å ÏùòÎØ∏: 2Ï∞®Ïõê ÌèâÎ©¥Ïùò ÏßÅÏÑ† Í∏ãÍ∏∞Î•º ÎÑòÏñ¥, Í≥†Ï∞®Ïõê Í≥µÍ∞ÑÏóê Ï¥àÌèâÎ©¥(Hyperplane)ÏùÑ ÎùÑÏö∞Îäî ÏûëÏóÖÏù¥Îã§.
3.  Í∞ÄÏ§ëÏπò($w$)Ïùò Ìï¥ÏÑù: "Îã§Î•∏ Î™®Îì† Ï°∞Í±¥Ïù¥ ÎèôÏùºÌï† Îïå(Ceteris Paribus)", Ìï¥Îãπ Î≥ÄÏàò ÌïòÎÇòÍ∞Ä Í≤∞Í≥ºÏóê ÎØ∏ÏπòÎäî ÏàúÏàòÌïú ÏòÅÌñ•Î†•ÏùÑ ÏùòÎØ∏ÌïúÎã§. Ïù¥Î•º ÌÜµÌï¥ ÏßàÎ≥ëÏùò ÌïµÏã¨ ÏúÑÌóò Ïù∏ÏûêÎ•º Ï∞æÏïÑÎÇº Ïàò ÏûàÎã§.
4.  Îã§Ï§ëÍ≥µÏÑ†ÏÑ±: ÎÑàÎ¨¥ ÎπÑÏä∑Ìïú(Ï§ëÎ≥µÎêú) Î≥ÄÏàòÎ•º ÎßéÏù¥ ÎÑ£ÏúºÎ©¥ Î™®Îç∏Ïù¥ ÌòºÎûÄÏùÑ Í≤™Í≥† Ìï¥ÏÑùÏù¥ ÏôúÍ≥°ÎêúÎã§. Î≥ÄÏàò ÏÑ†ÌÉùÏù¥ Ï§ëÏöîÌïòÎã§.

## 3.2. \[ÏòàÏ†ú 1\] Î≥ÄÏàòÍ∞Ä 2Í∞úÏùº ÎïåÏùò Í∏∞Ï¥à Î¨∏Î≤ï

ÏûÖÎ†• Îç∞Ïù¥ÌÑ∞($X$)Í∞Ä Ïù¥Ï†ú 1Ï§ÑÏù¥ ÏïÑÎãàÎùº Ïó¨Îü¨ Ï§Ñ(Ïª¨Îüº)ÏùÑ Í∞ÄÏßÑÎã§Îäî Ï†êÏùÑ
ÏΩîÎìúÎ°ú Ïù¥Ìï¥ÌïúÎã§.

ÏÉÅÌô©: ÌôòÏûêÏùò ÎÇòÏù¥(Age)ÏôÄ Ï≤¥Ï§ë(Weight) Îëê Í∞ÄÏßÄ Ï†ïÎ≥¥Î•º Ïù¥Ïö©Ìï¥
ÏàòÏ∂ïÍ∏∞ ÌòàÏïï(SBP)ÏùÑ ÏòàÏ∏°ÌïúÎã§.

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# 1. Îç∞Ïù¥ÌÑ∞ Ï§ÄÎπÑ (3Î™ÖÏùò ÌôòÏûê)
# X: [ÎÇòÏù¥, Ï≤¥Ï§ë] -> 2Í∞úÏùò ÌäπÏÑ±(Feature)
X = np.array([
    [25, 60],  # ÌôòÏûê A
    [45, 75],  # ÌôòÏûê B
    [65, 85]   # ÌôòÏûê C
])
# y: ÌòàÏïï
y = np.array([115, 130, 145])

# 2. Î™®Îç∏ ÌïôÏäµ
# Î¨∏Î≤ïÏùÄ Îã®Ïàú ÌöåÍ∑ÄÏôÄ 100% ÎèôÏùºÌïòÎã§. (Scikit-learnÏùò Ïû•Ï†ê)
model = LinearRegression()
model.fit(X, y)

# 3. ÌïôÏäµÎêú ÌååÎùºÎØ∏ÌÑ∞ ÌôïÏù∏
print(f"Í∞ÄÏ§ëÏπò(w1, w2): {model.coef_}")
print(f"Ï†àÌé∏(b): {model.intercept_:.2f}")

# 4. ÏàòÏãù Ìï¥ÏÑù
w_age, w_weight = model.coef_
print(f"Í≤∞Î°†Ïãù: ÌòàÏïï = {w_age:.2f}*ÎÇòÏù¥ + {w_weight:.2f}*Ï≤¥Ï§ë + {model.intercept_:.2f}")

# ÏÑ§Î™Ö: coef_Í∞Ä Î¶¨Ïä§Ìä∏ ÌòïÌÉúÎ°ú [ÎÇòÏù¥ Í∞ÄÏ§ëÏπò, Ï≤¥Ï§ë Í∞ÄÏ§ëÏπò] 2Í∞úÎ•º Î∞òÌôòÌïúÎã§.
```

> üìñ Î¨∏Ï†ú ÌíÄÏù¥ Í∞ÄÏù¥Îìú
>
> Ïù¥ ÏòàÏ†úÎäî ÏûÖÎ†• Î≥ÄÏàòÍ∞Ä 2Í∞ú Ïù¥ÏÉÅÏù∏ Îã§Ï§ë ÏÑ†Ìòï ÌöåÍ∑Ä(Multiple Linear Regression)Ïùò Í∏∞Ï¥àÎ•º Îã§Î£πÎãàÎã§.
> - ÌïµÏã¨ Î≥ÄÌôî: ÏûÖÎ†• Îç∞Ïù¥ÌÑ∞ $X$Í∞Ä Ïù¥Ï†ú '1Ï§Ñ(Vector)'Ïù¥ ÏïÑÎãàÎùº 'Ìëú(Matrix)' ÌòïÌÉúÍ∞Ä Îê©ÎãàÎã§.
> - Îç∞Ïù¥ÌÑ∞ Íµ¨Ï°∞: `[[ÎÇòÏù¥1, Ï≤¥Ï§ë1], [ÎÇòÏù¥2, Ï≤¥Ï§ë2], ...]` (NÌñâ 2Ïó¥)
> - ÏàòÏãù: $y = w_1 	imes 	ext{ÎÇòÏù¥} + w_2 	imes 	ext{Ï≤¥Ï§ë} + b$
> - Ìï¥ÏÑù: Î™®Îç∏ÏùÄ ÎÇòÏù¥ÏôÄ Ï≤¥Ï§ëÏù¥ ÌòàÏïïÏóê ÎØ∏ÏπòÎäî ÏòÅÌñ•ÏùÑ Í∞ÅÍ∞Å Î≥ÑÎèÑÏùò Í∞ÄÏ§ëÏπò($w_1, w_2$)Î°ú ÌïôÏäµÌï©ÎãàÎã§.



````python
#do write code
import numpy as np
from sklearn.linear_model import LinearRegression

# 1. Îç∞Ïù¥ÌÑ∞ Ï§ÄÎπÑ (3Î™ÖÏùò ÌôòÏûê)
# X: [ÎÇòÏù¥, Ï≤¥Ï§ë] -> 2Í∞úÏùò ÌäπÏÑ±(Feature)
X = np.array([
    [25, 60],  # ÌôòÏûê A
    [45, 75],  # ÌôòÏûê B
    [65, 85]   # ÌôòÏûê C
])
# y: ÌòàÏïï
y = np.array([115, 130, 145])

# 2. Î™®Îç∏ ÌïôÏäµ
# Î¨∏Î≤ïÏùÄ Îã®Ïàú ÌöåÍ∑ÄÏôÄ 100% ÎèôÏùºÌïòÎã§. (Scikit-learnÏùò Ïû•Ï†ê)
model = LinearRegression()
model.fit(X, y)

# 3. ÌïôÏäµÎêú ÌååÎùºÎØ∏ÌÑ∞ ÌôïÏù∏
print(f"Í∞ÄÏ§ëÏπò(w1, w2): {model.coef_}")
print(f"Ï†àÌé∏(b): {model.intercept_:.2f}")

# 4. ÏàòÏãù Ìï¥ÏÑù
w_age, w_weight = model.coef_
print(f"Í≤∞Î°†Ïãù: ÌòàÏïï = {w_age:.2f}*ÎÇòÏù¥ + {w_weight:.2f}*Ï≤¥Ï§ë + {model.intercept_:.2f}")
````

## 3.3. \[ÏòàÏ†ú 2\] ÏòàÏ∏° ÏÑ±Îä• ÎπÑÍµê: Ï†ïÎ≥¥Í∞Ä ÎßéÏùÑÏàòÎ°ù ÎòëÎòëÌï¥ÏßàÍπå?

2ÍµêÏãúÏóê ÌñàÎçò ÎãπÎá®Î≥ë Îç∞Ïù¥ÌÑ∞ÏÖãÏúºÎ°ú ÎèåÏïÑÍ∞ÄÏÑú, Î≥ÄÏàò 1Í∞úÎ•º Ïì∏ ÎïåÏôÄ 10Í∞ú
Ï†ÑÎ∂ÄÎ•º Ïì∏ ÎïåÏùò ÏÑ±Îä•ÏùÑ ÏßÅÏ†ë ÎπÑÍµêÌïúÎã§.

ÏÉÅÌô©: BMIÎßå ÏçºÏùÑ Îïå vs Ï†ÑÏ≤¥ Í≤ÄÏÇ¨ Í≤∞Í≥º(10Ï¢Ö)Î•º Îã§ ÏçºÏùÑ Îïå Ïò§Ï∞® ÎπÑÍµê.

```python
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# 1. Îç∞Ïù¥ÌÑ∞ Î°úÎìú
diabetes = load_diabetes()
X_all = diabetes.data       # 10Í∞ú ÌäπÏÑ± Î™®Îëê ÏÇ¨Ïö©
y = diabetes.target

# 2. Îç∞Ïù¥ÌÑ∞ Î∂ÑÌï†
X_train, X_test, y_train, y_test = train_test_split(X_all, y, test_size=0.2, random_state=42)

# 3. Î™®Îç∏ ÌïôÏäµ (Îã§Ï§ë ÌöåÍ∑Ä)
model_multi = LinearRegression()
model_multi.fit(X_train, y_train)

# 4. ÏÑ±Îä• ÌèâÍ∞Ä
y_pred = model_multi.predict(X_test)

r2 = r2_score(y_test, y_pred)

print("=== Îã§Ï§ë ÏÑ†Ìòï ÌöåÍ∑Ä ÏÑ±Ï†ÅÌëú ===")
print(f"MSE: {mse:.2f}")
print(f"R2 Score: {r2:.2f}")

# [ÎπÑÍµê] 2ÍµêÏãúÏùò Îã®Ïàú ÌöåÍ∑Ä Í≤∞Í≥º(BMIÎßå ÏÇ¨Ïö©)Îäî R2Í∞Ä ÏïΩ 0.23Ïù¥ÏóàÎã§.
# Îã§Ï§ë ÌöåÍ∑ÄÎäî R2Í∞Ä ÏïΩ 0.45~0.50 ÏàòÏ§ÄÏúºÎ°ú ÎåÄÌè≠ ÏÉÅÏäπÌïúÎã§.
# Í≤∞Î°†: ÌôòÏûêÏóê ÎåÄÌïú Ï†ïÎ≥¥Í∞Ä ÎßéÏùÑÏàòÎ°ù ÏòàÏ∏°Ïùò Ï†ïÌôïÎèÑÎäî Ïò¨ÎùºÍ∞ÑÎã§.
```


> üìñ Î¨∏Ï†ú ÌíÄÏù¥ Í∞ÄÏù¥Îìú
>
> Ï†ïÎ≥¥Í∞Ä ÎßéÏùÑÏàòÎ°ù ÏòàÏ∏°Î†•Ïù¥ Ï¢ãÏïÑÏßÄÎäîÏßÄ ÌôïÏù∏ÌïòÎäî Ïã§ÌóòÏûÖÎãàÎã§.
> - Ïã§Ìóò ÏÑ§Í≥Ñ:
>   - Case 1: Îã®Ïàú ÌöåÍ∑Ä (BMI ÌïòÎÇòÎßå ÏÇ¨Ïö©) -> 2ÍµêÏãú Í≤∞Í≥º Recall
>   - Case 2: Îã§Ï§ë ÌöåÍ∑Ä (ÎÇòÏù¥, ÏÑ±Î≥Ñ, ÌòàÏïï, ÌîºÍ≤ÄÏÇ¨ Îì± 10Í∞ú Î™®Îëê ÏÇ¨Ïö©)
> - ÌèâÍ∞Ä ÏßÄÌëú:
>   - R2 Score: 1Ïóê Í∞ÄÍπåÏö∏ÏàòÎ°ù Ï¢ãÏäµÎãàÎã§. Î≥ÄÏàòÍ∞Ä ÎäòÏñ¥ÎÇòÎ©¥ÏÑú Ïù¥ Ï†êÏàòÍ∞Ä ÏñºÎßàÎÇò Ïò§Î•¥ÎäîÏßÄ ÌôïÏù∏Ìï¥Î¥ÖÎãàÎã§.



````python
#do write code
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

diabetes = load_diabetes()
x_all = diabetes.data
y = diabetes.target

x_train, x_test, y_train, y_test = train_test_split(x_all, y, test_size=0.2, random_state=42)

model_multi = LinearRegression()
model_multi.fit(x_train, y_train)

y_pred = model_multi.predict(x_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("=== Îã§Ï§ë ÏÑ†Ìòï ÌöåÍ∑Ä ÏÑ±Ï†ÅÌëú ===")
print(f"MSE: {mse:.2f}")
print(f"R2 Score: {r2:.2f}")
````

## 3.4. \[ÏòàÏ†ú 3\] Ìó¨Ïä§ÏºÄÏñ¥Ïùò ÌïµÏã¨: Ï§ëÏöî Î≥ÄÏàò Ï∂îÏ∂ú (Feature Importance)

Îã®ÏàúÌûà ‚ÄúÏòàÏ∏°ÏùÑ ÏûòÌñàÎã§‚ÄùÎ°ú ÎÅùÎÇ¥Î©¥ Ïïà ÎêúÎã§. ‚ÄúÍ∑∏ÎûòÏÑú ÎãπÎá®Î≥ëÏùò Ï£ºÎ≤îÏù¥
Î≠îÎç∞?‚ÄùÎùºÎäî ÏùòÏÇ¨Ïùò ÏßàÎ¨∏Ïóê ÎãµÌï¥Ïïº ÌïúÎã§. $w$Í∞íÏùò ÌÅ¨Í∏∞Î•º ÎπÑÍµêÌïòÏó¨ Ï§ëÏöîÎèÑÎ•º
ÌååÏïÖÌïúÎã§.

ÏÉÅÌô©: 10Í∞úÏùò Í±¥Í∞ï ÏßÄÌëú Ï§ë ÎãπÎá®Î≥ë ÏïÖÌôîÏóê Í∞ÄÏû• ÌÅ∞ ÏòÅÌñ•ÏùÑ Ï£ºÎäî Top 3
ÏöîÏù∏ Ï∞æÍ∏∞.

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. ÌäπÏÑ± Ïù¥Î¶Ñ Í∞ÄÏ†∏Ïò§Í∏∞
feature_names = diabetes.feature_names
coefficients = model_multi.coef_

# 2. Î≥¥Í∏∞ Ï¢ãÍ≤å Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏúºÎ°ú Ï†ïÎ¶¨
df_coef = pd.DataFrame({
    'Feature': feature_names,
    'Coefficient (Weight)': coefficients
})

# 3. Í∞ÄÏ§ëÏπò Ï†àÎåìÍ∞í Í∏∞Ï§ÄÏúºÎ°ú ÎÇ¥Î¶ºÏ∞®Ïàú Ï†ïÎ†¨ (ÏòÅÌñ•Î†•Ïù¥ ÌÅ∞ ÏàúÏÑúÎåÄÎ°ú)
# Î∂ÄÌò∏(+/-)Îäî Î∞©Ìñ•ÏùÑ ÏùòÎØ∏ÌïòÍ≥†, Ï†àÎåìÍ∞íÏùÄ ÏòÅÌñ•Î†•Ïùò ÌÅ¨Í∏∞Î•º ÏùòÎØ∏ÌïúÎã§.
df_coef['Abs_Coef'] = df_coef['Coefficient (Weight)'].abs()

print("=== ÎãπÎá®Î≥ë ÏòÅÌñ• ÏöîÏù∏ ÏàúÏúÑ ===")
display(df_coef_sorted)

# 4. ÏãúÍ∞ÅÌôî
plt.figure(figsize=(10, 6))
# ÎßâÎåÄ Í∑∏ÎûòÌîÑÎ°ú Í∑∏Î¶¨Í∏∞
plt.barh(df_coef_sorted['Feature'], df_coef_sorted['Coefficient (Weight)'])
plt.xlabel('Coefficient Value')
plt.ylabel('Features')
plt.title('Feature Importance (Weights) in Diabetes Prediction')
plt.axvline(x=0, color='gray', linestyle='--', linewidth=0.8) # 0 Í∏∞Ï§ÄÏÑ†
```

> - ÏïÑÏù¥ÎîîÏñ¥: ÏÑ†Ìòï ÌöåÍ∑Ä Î™®Îç∏Ïù¥ ÌïôÏäµÌïú Í∞ÄÏ§ëÏπò($w$, Coefficient)Ïùò Ï†àÎåìÍ∞íÏù¥ ÌÅ¥ÏàòÎ°ù, Ìï¥Îãπ Î≥ÄÏàòÍ∞Ä Í≤∞Í≥º(ÎãπÎá®Î≥ë)Ïóê ÎØ∏ÏπòÎäî ÏòÅÌñ•Ïù¥ ÌÅΩÎãàÎã§.
