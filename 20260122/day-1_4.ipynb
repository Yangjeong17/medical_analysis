import numpy as np

## 4.6. \[ì˜ˆì œ 5\] \[ì¤‘ìš”\] ì‹¤ì „ íŒŒì´í”„ë¼ì¸: ë‚˜ëˆ„ê³ (Split), ì ìš©í•˜ë¼!

from sklearn.model_selection import train_test_split
