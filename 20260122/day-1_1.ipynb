## 1. 패러다임의 대전환: 연역법에서 귀납법으로

우리가 흔히 '컴퓨터 프로그래밍'이라고 부르는 것과 '머신러닝'은 문제를 해결하는 사고방식 자체가 완전히 반대이다. 이를 논리학의 용어를 빌려 설명하면 연역법과 귀납법의 차이와 같다.

### 1) 전통적 프로그래밍: "정답을 알려줄게, 계산은 네가 해" (연역적 접근)
지금까지 인류가 컴퓨터를 써온 방식은 '명령(Command)'이었습니다. 사람이 세상의 이치를 이미 다 알고 있고, 이를 컴퓨터 언어로 번역해서 입력해주는 방식이다.

*   상황: 의사가 고열 환자를 보고 독감인지 아닌지 판별하는 프로그램을 만든다고 가정해 봅시다.
*   접근 방식: 의사는 자신의 지식과 경험을 동원해서 '명시적인 규칙(Explicit Rule)'을 만듭니다.
    > "체온이 38도 이상이고, 기침을 하며, 오한이 있으면 $\rightarrow$ 독감이다."
    > "체온이 37.5도 이하이고, 콧물만 나면 $\rightarrow$ 일반 감기다."
*   컴퓨터의 역할: 컴퓨터는 아무런 생각이 없습니다. 그저 의사가 짜준 `if (체온 > 38) ...` 코드를 한 치의 오차도 없이, 빛의 속도로 실행하는 고성능 계산기일 뿐이다.
*   구조: `데이터(Input)` + `규칙(Rule)` $\rightarrow$ `정답(Output)`

### 2) 머신러닝: "예시를 보여줄게, 규칙은 네가 찾아" (귀납적 접근)
하지만 세상은 그렇게 단순하지 않다. 의학은 예외투성이다. 체온이 37도인데 독감인 사람도 있고, 39도인데 독감이 아닌 사람도 있다. 이 수만 가지 예외를 사람이 일일이 `if-else` 문으로 짜 넣을 수 있는가? 불가능하다. 새로운 방법론 등장이 필요하다.

*   접근 방식: 의사는 규칙을 만드는 대신, 과거의 '데이터(Data)'를 모아서 컴퓨터에게 던져줍니다.
    > "자, 여기 환자 10,000명의 진료 기록이 있어. 이 사람은 열이 38.2도였는데 독감이었어. 저 사람은 37.5도였는데 아니었어. 정답지(진단 결과)도 같이 줄게."
    > "이 데이터를 보고 독감인지 아닌지 나누는 기준은 네가 알아서 찾아봐."
*   컴퓨터의 역할: 컴퓨터는 이제 계산기가 아니라 학생이 됩니다. 데이터를 훑어보며 공통점을 찾고 차이점을 발견한다. 수많은 시행착오 끝에 "아하! 대략 체온이 37.8도 근처에서 기침 빈도가 5회 이상이면 독감 확률이 높구나"라는 패턴을 스스로 깨닫는다.
*   구조: `데이터(Input)` + `정답(Output)` $\rightarrow$ `규칙(Rule)`

이것이 머신러닝의 핵심이다. 데이터로부터 규칙을 학습(Learning)하는 것. 사람이 짜준 코드가 아니라, 데이터가 만들어낸 코드(모델)로 세상을 판단하는 것이다.

## 2. 왜 지금 머신러닝인가? (헬스케어의 한계 돌파)

왜 우리는 굳이 기계에게 학습을 시키려 할까요? 사람이 직접 규칙을 짜는 게 더 정확하지 않을까요? 헬스케어 분야의 예시를 통해 그 이유를 알아봅니다.

### 1) 규칙의 복잡성: 사람이 감당할 수 없는 변수의 수
'패혈증(Sepsis)'을 조기에 진단하는 문제를 생각해 봅시다. 패혈증은 시간 싸움입니다.
*   규칙 기반 접근: 의사들은 'SIRS 기준'이나 'qSOFA 점수' 같은 가이드라인을 만들었습니다. 호흡수 22회 이상, 혈압 100 이하 등 몇 가지 기준을 체크합니다. 하지만 이것만으로는 수많은 위양성(가짜 경보)과 위음성(놓침)이 발생합니다.
*   문제점: 사람의 몸은 너무나 복잡합니다. 혈압, 맥박, 체온뿐만 아니라 나이, 기저질환, 복용 약물, 최근 수술 이력, 유전자 정보 등 수백 가지 변수가 복합적으로 작용합니다. 사람의 머리로는 변수 3~4개를 동시에 고려하기도 벅찹니다. 100개의 변수가 얽히고설킨 관계를 수식으로 정의하는 건 인간 지능의 한계를 넘어서는 일입니다.
*   머신러닝의 해결: 머신러닝은 100차원, 1000차원의 데이터를 동시에 처리할 수 있습니다. "나이가 많으면서 혈압은 낮지만 백혈구 수치가 급격히 오르는 패턴" 같은 아주 복잡하고 미세한 조합을 찾아낼 수 있습니다.

### 2) 암묵지(Tacit Knowledge)의 한계: 말로 설명할 수 없는 노하우
영상의학과 명의는 엑스레이 사진을 딱 보자마자 "어? 이거 느낌이 쎄한데?" 하고 암을 찾아냅니다. 하지만 후배 의사가 "교수님, 왜 거기가 암이라고 생각하셨습니까?"라고 물으면 설명하기 어려울 때가 많다.
*   "음... 여기가 약간 거칠고, 음영이 미묘하게 다르잖아."
*   이것은 언어나 수식으로 정의하기 힘든 '직관' 혹은 '암묵지'의 영역입니다. 설명할 수 없으니 코드로 짤 수도 없다. (`if 모양 == 쎄하다?` $\rightarrow$ 코딩 불가)
*   머신러닝(특히 딥러닝)은 이미지 데이터를 픽셀 단위로 분석하여, 명의조차 말로 설명하지 못했던 그 '미묘한 패턴'을 수치적으로 학습해낸다.

### 3) 변화하는 환경: 데이터 드리프트(Data Drift)
의학 지식은 변합니다. 새로운 바이러스(예: 코로나19)가 등장하면 기존의 폐렴 진단 규칙은 무용지물이 될 수 있다.
*   규칙 기반: 새로운 바이러스의 특징을 연구해서, 사람이 다시 코드를 수정하고 배포해야 합니다. 시간이 오래 걸린다.
*   머신러닝: 새로운 환자 데이터가 쌓이면, AI는 그것을 다시 학습(Retraining)하기만 하면 됩니다. 변화하는 환경에 맞춰 규칙을 스스로 업데이트한다.

## 3. 머신러닝의 학습 과정: 갓난아기 비유

머신러닝 모델이 학습하는 과정은 갓난아기가 세상을 배우는 과정과 놀랍도록 닮았다. 이를 '사과와 배 구분하기'로 비유해 보자.



### 1단계: 모델 초기화 (백지상태)
아기(모델)는 태어나서 처음 과일을 본다. 사과가 뭔지, 배가 뭔지 전혀 모릅니다. 뇌(파라미터)가 백지상태이거나 무작위로 설정되어 있습니다.


### 2단계: 예측 (Inference)
엄마가 빨간 사과를 보여주며 묻습니다. "이게 뭐게?"
아기는 아무것도 모르니까 그냥 찍습니다. "배!"
*   이것이 모델의 초기 예측입니다. 당연히 틀릴 확률이 높습니다.



### 3단계: 오차 계산 (Loss Function)
엄마가 정답을 알려줍니다. "땡! 이건 사과야."
아기는 자신의 대답(배)과 정답(사과) 사이의 차이, 즉 '오차(Error)'를 인지합니다. 머신러닝에서는 이를 손실(Loss) 또는 비용(Cost)이라고 부릅니다. 학습의 목표는 이 '손실'을 '0'으로 만드는 것입니다.

### 4단계: 가중치 업데이트 (Optimization)
아기는 생각합니다. '아, 내가 색깔을 안 보고 모양만 봐서 틀렸구나. 다음엔 색깔을 더 중요하게 봐야지.'
뇌 속의 뉴런 연결 강도(가중치, Weight)를 조절합니다.
*   "색깔(빨강)"에 대한 가중치 $\uparrow$ (중요도 높임)
*   "모양(동그라미)"에 대한 가중치 $\downarrow$ (중요도 낮춤, 배도 동그라니까 구분에 도움 안 됨)

### 5단계: 반복 (Iteration)
엄마는 계속해서 다른 사과, 초록 아오리 사과, 노란 배, 갈색 배를 보여줍니다.
아기는 예측하고 $\rightarrow$ 틀리고 $\rightarrow$ 혼나고(오차) $\rightarrow$ 수정하는 과정을 수천, 수만 번 반복합니다.
이 과정을 통해 아기의 뇌 속에는 "빨갛고 매끄러우면 사과", "노랗고 점이 박혀 있으면 배"라는 정교한 판별식이 생겨납니다. 이것이 학습 완료된 모델입니다.


우리 두뇌는 뉴런이 시넵스를 통해 전기 신호를 주고받는 구조로 되어있다. 뉴런이란 인간의 뇌 속에서 신경전달물질을 통해 신호를 전달하고 정보를 받아들이며 처리하는 역할을 하는 물질로, 인간은 약 600억개에서 1000억개의 뉴런(Neuron)을 가지고 태어난다고 한다. 또한 이러한 뉴런에는 자극을 세포 밖으로 전도시키는 돌기인 축삭이라는 부분이 있는데, 축삭의 끝부분과 신경전달물질이 오가는 다음 뉴런 사이의 틈을 시냅스(Synaps)라고 한다. (출처:서울아산병원)

## 4. 머신러닝을 위한 필수 용어 정리

이제 본격적인 코딩에 앞서, 데이터 분석가들이 사용하는 언어를 정리해보자. 헬스케어 데이터를 예로 들어 매칭해 본다.

| 머신러닝 용어 | 헬스케어 데이터 매칭 | 의미 |
| :--- | :--- | :--- |
| 샘플 (Sample / Instance) | 환자 1명 | 데이터의 한 행(Row). 분석의 대상이 되는 개별 단위. |
| 특성 (Feature / Attribute) | 검사 항목 ($x$) | 나이, 성별, 혈압, BMI, 혈당 수치 등. 정답을 맞히기 위한 단서(Hint). 수학적으로는 $x$로 표기. |
| 레이블 (Label / Target) | 진단 결과 ($y$) | 암 여부(양성/악성), 생존 여부, 혈압 수치 등. 모델이 맞춰야 하는 정답. 수학적으로는 $y$로 표기. |
| 모델 (Model) | 진단 알고리즘 | 입력($x$)을 받아 출력($y$)을 뱉어내는 수학적 함수 박스. ($y = f(x)$) |
| 학습 (Training / Fitting) | 과거 차트 공부 | 모델에게 문제($x$)와 정답($y$)을 보여주며 최적의 함수 $f$를 찾는 과정. |
| 예측 (Prediction / Inference) | 신규 환자 진단 | 학습된 모델에 새로운 환자의 $x$만 넣어서 $y$를 추측하게 하는 것. |

### 데이터셋의 분할: 모의고사와 수능
우리가 100명의 환자 데이터(문제+정답)를 가지고 있다고 칩시다. 이 100명을 몽땅 털어서 모델을 학습시키면 어떻게 될까요?
모델은 100명의 환자를 달달 외워버립니다. 그리고 "학습 끝났어요!"라고 외칩니다.
선생님이 그 100명 중 한 명을 데려와서 테스트를 하면 100점을 맞겠죠. 하지만 이게 진짜 실력일까요? 아닙니다. 그냥 답안지를 외운 겁니다.
내일 병원에 처음 오는 '101번째 새로운 환자'도 맞출 수 있느냐가 중요합니다.

그래서 우리는 데이터를 반드시 두 덩어리로 나눕니다.
1.  훈련 데이터 (Training Set): 교과서. 모델이 공부하는 데 씁니다. (보통 70~80%)
2.  테스트 데이터 (Test Set): 수능 시험지. 모델이 공부할 때는 절대로 보여주지 않고 숨겨둡니다. 나중에 모델의 실력을 검증할 때만 딱 한 번 꺼내 봅니다. (보통 20~30%)

## 5. 정리: 머신러닝은 마법이 아니다

머신러닝은 인공지능이 스스로 자아를 가지고 생각하는 마법이 아니다.
그것은 "데이터라는 재료를 믹서기에 넣고 갈아서, 그 안에 숨겨진 통계적 패턴(수학 공식)을 찾아내는 아주 부지런한 과정"일 뿐이다.

*   우리가 할 일은 좋은 재료(Clean Data)를 준비하고,
*   적절한 믹서기(Model Algorithm)를 고르고,
*   믹서기가 잘 돌아가는지 감시(Evaluation)하는 것이다.

# 📘 \[1교시\] 머신러닝의 본질과 Scikit-learn의 기초

## 1.1. \[개념 심화\] 머신러닝: 규칙 기반에서 데이터 기반으로

### 1) 패러다임의 전환: 연역법에서 귀납법으로

우리가 지금까지 해왔던 전통적인 프로그래밍과 머신러닝은 문제를 해결하는
접근 방식 자체가 다르다. 의료 진단을 예로 들어 설명한다.

- **전통적 프로그래밍 (Rule-based):**의사나 프로그래머가 **‘명시적인
  규칙’**을 코딩한다.
  - *논리:* “만약 체온이 37.5도 이상이고, 기침을 하면 $\rightarrow$
    감기라고 판정하라.”
  - *한계:* 인체의 생리는 너무나 복잡하다. 패혈증(Sepsis)을 진단하기
    위해 고려해야 할 변수는 수십, 수백 가지다. 이를 `if-else` 문으로
    모두 작성하는 것은 불가능하다.
- **머신러닝 (Data-driven):** 기계에게 **‘데이터(문제)’**와
  **‘정답(결과)’**을 던져주고, **‘규칙’**을 스스로 찾게 한다.
  - *논리:* “여기 환자 10,000명의 기록(체온, 혈압, 나이…)과
    결과(생존/사망)가 있다. 어떤 패턴일 때 사망 위험이 높은지 네가
    수식을 찾아내라.”
  - *장점:* 인간이 인지하지 못하는 미세한 패턴이나 비선형적인 관계까지
    찾아낼 수 있다.

### 2) 학습(Learning)의 수학적 의미

머신러닝에서 ’학습한다’는 것은 추상적인 표현이 아니다. 이는 **$y = f(x)$
라는 함수에서, 입력 $x$와 출력 $y$를 가장 잘 연결해주는 함수 $f$의
모양(파라미터)을 찾는 과정**이다. \* **입력($x$):** 환자의 데이터 (특성,
Feature) \* **출력($y$):** 진단 결과 (타겟, Label/Target) \*
**모델($f$):** 기계가 만들어낸 판단 규칙

### 3) Scikit-learn의 3단계 철학 (The Estimator API)

파이썬 머신러닝 생태계의 표준인 Scikit-learn은 어떤 복잡한 알고리즘이든
다음 3단계로 통일하여 사용한다. 이를 **‘일관성 있는 인터페이스’**라고
한다. 1. **모델 객체 생성 (`Model()`):** “어떤 방법으로 학습할래?” (예:
선을 그을래? 스무고개를 할래?) 2. **학습 (`.fit(X, y)`):** “여기
문제($X$)와 정답($y$)이 있으니 공부해.” $\rightarrow$ 데이터의 패턴을
파악하여 모델 내부의 파라미터를 업데이트한다. 3. **예측
(`.predict(X_new)`):** “이제 새로운 환자($X_{new}$)가 왔으니 결과를
맞춰봐.” $\rightarrow$ 학습된 규칙을 적용한다.



머신러닝 실습(특히 Scikit-learn 및 NumPy 활용)에서 가장 빈번하게 사용되는 핵심 함수들의 API 명세서이다.

### 💡 [Tip] 자주 발생하는 에러와 해결책

1.  **ValueError: Expected 2D array, got 1D array instead**
    *   **원인:** 모델 학습(`fit`)이나 예측(`predict`) 시 입력 데이터 $X$는 반드시 **행렬(2차원)** 형태여야 하는데, 리스트나 1차원 배열을 넣음.
    *   **해결:** `.reshape(-1, 1)`을 사용하여 2차원으로 만들어줌.
    *   *예:* `model.predict([30])` (X) $\rightarrow$ `model.predict([[30]])` (O)

2.  **Data Leakage (데이터 누수)**
    *   **원인:** 전처리(Scaler)를 할 때 `fit`을 전체 데이터나 Test 데이터에 적용함.
    *   **해결:** `fit`은 무조건 **Train Data**에만 수행하고, Test Data는 `transform`만 수행함.

## 1.2. \[예제 1\] 가장 단순한 관계: 체온과 심박수 (1:1 관계)

가장 기초적인 1차원 데이터를 통해 `fit`과 `predict`의 흐름을 익힌다.

**상황:** 체온이 오르면 심박수도 오르는 경향이 있다. \* 데이터: 3명의
환자 기록.

``` python
import numpy as np
from sklearn.linear_model import LinearRegression

# 1. 데이터 준비 (x: 체온, y: 심박수)
# 머신러닝에서 입력 데이터(X)는 항상 '행렬(2차원 배열)' 형태여야 한다.
# reshape(-1, 1)은 데이터를 세로로 길게 세우는 역할을 한다.
X_temp = np.array([36.5, 37.0, 37.5]).reshape(-1, 1)
y_heart = np.array([70, 80, 90])

print("입력 데이터(X):\n", X_temp)
print("정답 데이터(y):", y_heart)

# 2. 모델 생성
model = LinearRegression()

# 3. 학습 (fit)
# 기계는 이 시점에 "체온이 0.5도 오를 때마다 심박수가 10씩 오르는구나"라는 규칙을 깨닫는다.
model.fit(X_temp, y_heart)

# 4. 예측 (predict)
# 체온이 38.0도인 새로운 환자가 왔다. 심박수는 얼마일까?
new_patient = np.array([[38.0]])
prediction = model.predict(new_patient)

print(f"체온이 38.0도인 환자의 예상 심박수: {prediction[0]:.1f} BPM")
```

> **📖 문제 풀이 가이드**
>
> 이 문제는 가장 기본적인 **선형 회귀(Linear Regression)** 모델을 만들어보는 예제입니다.
> - **목표**: 체온(X)과 심박수(y) 사이의 정비례 관계를 모델이 학습하도록 합니다.
> - **사용 함수**:
>   - `LinearRegression`: 데이터를 가장 잘 설명하는 직선을 찾는 모델입니다.
>   - `.fit(X, y)`: 모델에게 **문제(X)**와 **정답(y)**을 주어 데이터의 패턴을 **학습**시키는 핵심 함수입니다.
>   - `.predict(X)`: 학습된 모델에 새로운 문제(X)를 넣어 **정답을 예측**하는 함수입니다.


````python
#do write code here

import numpy as np
from sklearn.linear_model import LinearRegression

x_temp = np.array([36.5, 37.0, 37.5]).reshape(-1, 1)
y_heart = np.array([70, 80, 90])

print("입력 데이터(x):\n", x_temp)
print("정답 데이터(y):", y_heart)

model = LinearRegression()

model.fit(x_temp, y_heart)

new_patient = np.array([[38.0]])
prediction = model.predict(new_patient)

print(f"체온이 38.0도인 환자의 예상 심박수: {prediction[0]:.1f} BPM")
````

## 1.3. \[예제 2\] 데이터의 차원 이해: 약물 용량과 혈압 (Shape의 중요성)

많은 초심자가 데이터의 형태(Shape) 때문에 에러를 겪는다. 이를 명확히
짚고 넘어간다.

**상황:** 고혈압 약 투여 용량(mg)에 따른 혈압 감소량(mmHg).

``` python
# 데이터 준비
# X: 약물 용량 (mg)
# y: 혈압 감소량 (mmHg)
X_drug = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
y_drop = np.array([5, 12, 18, 25, 31]) # 대략 용량의 0.6배 + 약간의 오차

# 모델 학습
model_bp = LinearRegression()
model_bp.fit(X_drug, y_drop)

# 모델이 찾아낸 규칙(기울기) 확인
# coef_는 모델이 학습한 '가중치'를 의미한다.
print(f"학습된 규칙: 약물 1mg당 혈압은 약 {model_bp.coef_[0]:.2f} mmHg 감소한다.")

# 예측
# 60mg을 투여하면 얼마나 떨어질까?
pred_bp = model_bp.predict([[60]])
print(f"60mg 투여 시 예상 혈압 감소량: {pred_bp[0]:.2f} mmHg")
```

> **📖 문제 풀이 가이드**
>
> 이 문제는 데이터의 **형태(Shape)**를 맞추는 것이 핵심입니다. 머신러닝 모델은 입력 데이터의 차원에 민감하게 작동합니다.
> - **핵심 포인트**: Scikit-learn의 모델들은 입력 데이터($X$)로 **2차원 행렬(Matrix)**을 기대합니다.
>   - 1차원 배열: `[1, 2, 3]` (벡터) -> 모델이 이해하지 못함 ❌
>   - 2차원 배열: `[[1], [2], [3]]` (N행 1열의 행렬) -> 모델이 이해함 ✅
> - **사용 함수**:
>   - `.reshape(-1, 1)`: 배열의 모양을 변경합니다. `-1`은 '데이터 개수만큼 자동으로 행을 맞추라'는 의미이고, `1`은 '열을 1개로 만들라'는 의미입니다. 즉, 데이터를 세로로 길게 세우는 역할을 합니다.


````python
#do write code here
x_drug = np.array([10,20,30,40,50]).reshape(-1,1)
y_drop = np.array([5,12,18,25,31])

model_bp = LinearRegression()
model_bp.fit(x_drug, y_drop)

print(f"학습된 규칙: 약물 1mg당 혈압은 약 {model_bp.coef_[0]:.2f} mmHg 감소한다.")

pred_bp = model_bp.predict([[60]])
print(f"60mg 투여 시 예상 혈압 감소량: {pred_bp[0]:.2f} mmHg")
````

``` python
# [나이, 체중]
X_multi = np.array([
```

## 1.5. \[예제 4\] 회귀(Regression) vs 분류(Classification) 맛보기

지금까지는 숫자를 예측했다. 이번에는 질병 유무(0 또는 1)를 예측하는
논리를 간단히 경험한다. (상세 내용은 2일차에 배우지만, `fit`-`predict`
구조가 동일함을 보여준다.)

**상황:** 종양의 크기(cm)를 보고 악성(1)인지 양성(0)인지 판단한다.

``` python
from sklearn.linear_model import LogisticRegression

# 1. 데이터 준비
# 종양 크기
X_tumor = np.array([0.5, 1.0, 1.5, 3.5, 4.0, 5.0]).reshape(-1, 1)

y_cancer = np.array([0, 0, 0, 1, 1, 1])
```

> - **목표**: 종양의 크기를 보고 악성(1)인지 양성(0)인지 분류합니다.
> - **사용 함수**:
>   - `LogisticRegression`: 이름은 '회귀'지만, 실제로는 **분류** 문제에 주로 사용되는 확률 예측 모델입니다. 데이터가 특정 클래스(0 또는 1)에 속할 확률을 계산합니다.

````python
#do write code here
from sklearn.linear_model import LogisticRegression

x_tumor = np.array([0.5,1.0,1.5,3.5,4.0,5.0]).reshape(-1,1)
y_cancer = np.array([0,0,0,1,1,1])

clf = LogisticRegression()
clf.fit(x_tumor, y_cancer)

test_tumor = np.array([[3.0]])
result = clf.predict(test_tumor)

diagnosis = "악성(암)" if result[0] == 1 else "양성(정상)"
print(f"종양 크기 3.0cm 진단 결과: {diagnosis}")

#다른 버전

test_tumor = np.array([[3.0]])
result = clf.predict(test_tumor)


if result[0] == 1:
    print(f"종양 크기 3.0cm 진단결과: cancer")
else:
    print(f"종양 크기 3.0cm 진단 결과: normal")
````
